```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Loading clustered dataset
file = "Resources/MLM/clustered_data.csv"
clustered_df = pd.read_csv(file, index_col=0)
clustered_df.head(10)
```

```python
clustered_df.dtypes
```

```python
numerical_df = pd.get_dummies(data = clustered_df.drop(columns=['PC 1','PC 2','PC 3']),
                              columns=['abortion_status','prescription_contraception',
                              'otc_methods','male_sterilization','female_sterilization',
                              'cost_sharing'])
numerical_df
```

```python
numerical_df.dtypes
```

```python
numerical_df['chc_per_capita_land'] = ((numerical_df['total_community_health_centers'] / 
                                        numerical_df['population']) * 100000) / (numerical_df['land_area_sqmi']/100000)
numerical_df.head(10)
```

```python
# Dropping Total CHCs, population, and land area
dropped_df = numerical_df.drop(columns=['total_community_health_centers','population','land_area_sqmi'])
dropped_df.head()
```

```python
training_df = dropped_df.loc[(dropped_df['Class']==4) | (dropped_df['Class']==2)]
training_df
```

```python
# Dropping DC 
training_df = training_df.loc[training_df.index!='District of Columbia']
training_df
```

```python
X_train = training_df.drop(columns='chc_per_capita_land')
y_train = training_df[['chc_per_capita_land']]

y_train['chc_per_capita_land'].value_counts()
```

```python
fit_df = dropped_df.loc[(dropped_df['Class']!=4)&(dropped_df['Class']!=2)]
fit_df.head()
```

```python
X = fit_df.drop(columns='chc_per_capita_land')
y = fit_df[['chc_per_capita_land']]
```

```python
# Creating an object of LinearRegression class
LR = LinearRegression()

# Fitting the training data
LR.fit(X_train, y_train)
```

```python
y_pred = LR.predict(X)
y_pred
```

```python
y_pred.shape
```

```python
compare_df = y['chc_per_capita_land'].to_frame()
compare_df['predictions'] = y_pred
compare_df['population'] = numerical_df['population']
compare_df['land_area_sqmi'] = numerical_df['land_area_sqmi']
compare_df['chc_current'] = numerical_df['total_community_health_centers']
compare_df['chc_needed'] = compare_df['predictions'] * compare_df['population']/100000 * compare_df['land_area_sqmi']/100000

compare_df
```

```python
# Predicting the accuracy score
score = r2_score(y,y_pred)
print('r2 score is ',score)
print('mean_sqrd_error is ',mean_squared_error(y,y_pred))
print('root_mean_squared_error is ',np.sqrt(mean_squared_error(y,y_pred)))
```
