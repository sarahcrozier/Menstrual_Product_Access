# Machine Learning
### Input variables
### State abortion policy as of 8/7/22 
### State Community Health Centers
### Health insurance coverage of women
### Maternal death and mortality rates
### Legal abortions obtained out of state
### Legal abortions obtained in state of occurence
### Population distribution by sex
### Reported women not seeing a doctor in past 12 months
### Reported women having a mammogram in last 2 years
### Reported women not having a personal doctor or health care provider
### Reported women having a pap smear in the last 3 years
### State requirements for insurance coverage of contraceptives
### Poverty 

```python
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
abortion_policy_df = pd.read_csv('Abortion_Policy.csv', index_col='State')
abortion_policy_df = abortion_policy_df.rename(columns={"Status of Abortion": "abortion_status"})
abortion_policy_df
```

```python
CHC_df = pd.read_csv("Community_Health_Center_Delivery_Sites_and_Patient_Visits.csv", index_col='State')
CHC_df = CHC_df.rename(columns={"Total CHCs": "total_community_health_centers"})
CHC_df
```

```python
ins_df = pd.read_csv("Health_Ins.csv", index_col='State')
ins_df = ins_df.drop(columns=['Total','%_Uninsured','%_Insured'])
ins_df = ins_df.rename(columns={'Uninsured':'uninsured', "Insured":'insured'})
ins_df
```

```python
maternal_mortality_df = pd.read_csv("Maternal_Mortality.csv", index_col='State')
maternal_mortality_df = maternal_mortality_df.rename(columns={'Maternal_Mortality':"maternal_mortality"})
maternal_mortality_df
```

```python
population_df = pd.read_csv("Population.csv", index_col='State')
population_df = population_df.drop(columns=['%_total_population'])
population_df = population_df.rename(columns={"Population": "population"})
population_df
```

```python
no_doctor_visit_df = pd.read_csv("No_Doctor_Visit.csv", index_col='State')
no_doctor_visit_df = no_doctor_visit_df.rename(columns={"All Women": "no_doctor_visits"})
no_doctor_visit_df
```

```python
mammogram_df = pd.read_csv("Mammogram.csv", index_col='State')
mammogram_df = mammogram_df.rename(columns={"All Women": "mammogram"})
mammogram_df
```

```python
no_doctor_df = pd.read_csv("No_Provider.csv", index_col='State')
no_doctor_df = no_doctor_df.rename(columns={"All Women": "no_provider"})
no_doctor_df
```

```python
pap_smear_df = pd.read_csv("Pap_Smear.csv", index_col='State')
pap_smear_df = pap_smear_df.rename(columns={"All Women": "pap_smear"})
pap_smear_df
```

```python
ins_contraceptives_df = pd.read_csv("Ins_Coverage_Contraceptives.csv",index_col='State')
ins_contraceptives_df = ins_contraceptives_df.rename(columns={"State Requires Coverage of Prescription Contraception": "prescription_contraception", 
                                                             "State Requires Coverage of Over-the-counter Methods": "otc_methods",
                                                             "State Requires Coverage of Male Sterilization": "male_sterilization",
                                                             "State Requires Coverage of Female Sterilization": "female_sterilization",
                                                             "State Prohibits cost sharing": "cost_sharing"})

ins_contraceptives_df
```

```python
teen_birth_df = pd.read_csv("Teen_Birth_Rate.csv", index_col='State')
teen_birth_df = teen_birth_df.rename(columns={"Teen Birth Rate per 1,000": "teen_births"})
teen_birth_df
```

```python
poverty_df = pd.read_csv("poverty_pop_distribution.csv", index_col='State')
poverty_df = poverty_df.rename(columns={"Poverty_Under_200": "poverty_under_200"})
poverty_df
```

```python
#https://stackoverflow.com/questions/44327999/python-pandas-merge-multiple-dataframes
data_frames = [abortion_policy_df, CHC_df, ins_df, maternal_mortality_df, population_df, no_doctor_visit_df, 
              mammogram_df, no_doctor_df, pap_smear_df, ins_contraceptives_df, teen_birth_df, poverty_df]
```

```python
from functools import reduce
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['State'],
                                            how='outer'), data_frames)
```

```python
df_merged.index.names = ['state']
```

```python
df_merged.head(5)
```

```python
df_merged.dtypes
```

```python
pd.DataFrame.to_csv(df_merged, 'merged.csv', sep=',', na_rep='.', index=True)
```

```python
df_merged.to_sql(name='health_data', con=engine)
```
