```python
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
abortion_status_df = pd.read_csv("All_States_Abortion_Policy_Tracker.csv", 
                                        skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76,77,78,79,80],
                                        index_col='Location')
abortion_status_df = abortion_status_df.drop(columns='Footnotes')
abortion_status_df = abortion_status_df.rename(columns={"Status of Abortion": "Abortion_Status"})
abortion_status_df
```

```python
CHC_df = pd.read_csv("Community_Health_Center_Delivery_Sites_and_Patient_Visits.csv", 
                                        skiprows=[0,1,3,42,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76,77,78,79],
                                        index_col='Location')
CHC_df = CHC_df.drop(columns='Footnotes')
CHC_df = CHC_df.rename(columns={"Total CHCs": "Total_Community_Health_Centers"})
CHC_df
```

```python
ins_df = pd.read_csv("Health_Ins_Coverage_of_Women_19-64_in_2020.csv", 
                         skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76,77,78,79,80,81], 
                         index_col='Location')
ins_df = ins_df.drop(columns=['Employer', 'Non-Group', 'Medicaid', 'Medicare', 'Military'])
ins_df = ins_df.rename(columns={"Total":"Total_Insured"})
ins_df
```

```python
maternal_deaths_df = pd.read_csv("Maternal_deaths_and_mortality_rates.csv", 
                                    skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69], 
                                    index_col='Location')
maternal_deaths_df = maternal_deaths_df.drop(columns=['Maternal Mortality Rate per 100,000 live Births'])
maternal_deaths_df = maternal_deaths_df.rename(columns={"Number of Deaths": "Number Maternal Deaths"})
maternal_deaths_df
```

```python
total_births_df = pd.read_csv("Total_Number_of_Births.csv", 
                                    skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72], 
                                    index_col='Location')
total_births_df = total_births_df.drop(columns='Footnotes')
total_births_df['Number of Births'] = total_births_df['Number of Births'].str.replace(',', '')
total_births_df
```

```python
total_births_df['Number of Births'] =  pd.to_numeric(total_births_df['Number of Births'])
death_div_births = (maternal_deaths_df["Number Maternal Deaths"] / total_births_df["Number of Births"])
death_div_births_mult = death_div_births * 100000
maternal_mortality_df = pd.DataFrame(data=death_div_births_mult, columns=['Maternal_Mortality'])
maternal_mortality_df
```

```python
population_df = pd.read_csv("Population_Distribution_by_Sex.csv", 
                            skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69], 
                            index_col='Location')
population_df = population_df.drop(columns=['Footnotes', 'Male', 'Female'])
population_df = population_df.rename(columns={"Total":"Population"})
population_df
```

```python
no_doctor_visit_df = pd.read_csv("Report_Not_Seeing_Doctor_in_Past_12_Months_Due_to_Cost.csv", 
                                 skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75], 
                                 index_col='Location')
no_doctor_visit_df = no_doctor_visit_df.drop(columns='Footnotes')
no_doctor_visit_df = no_doctor_visit_df.rename(columns={"All Women": "No_Doctor_Visits"})
no_doctor_visit_df
```

```python
mammogram_df = pd.read_csv("Reported_Mammogram_2_Years.csv", 
                           skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73], 
                           index_col='Location')
mammogram_df = mammogram_df.drop(columns='Footnotes')
mammogram_df = mammogram_df.rename(columns={"All Women": "Mammogram"})
mammogram_df
```

```python
no_doctor_df = pd.read_csv("Reported_No_Personal_Doctor_or_Health_Care_Provider.csv", 
                           skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75], 
                           index_col='Location')
no_doctor_df = no_doctor_df.drop(columns='Footnotes')
no_doctor_df = no_doctor_df.rename(columns={"All Women": "No_Provider"})
no_doctor_df
```

```python
pap_smear_df = pd.read_csv("Reported_Pap_Smear_3_Years.csv", 
                           skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71], 
                           index_col='Location')
pap_smear_df = pap_smear_df.drop(columns='Footnotes')
pap_smear_df = pap_smear_df.rename(columns={"All Women": "Pap_Smear"})
pap_smear_df
```

```python
ins_contraceptives_df = pd.read_csv("State_Requirements_Ins_Coverage_Contraceptives.csv",
                                   skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74],
                                   index_col='Location')
ins_contraceptives_df = ins_contraceptives_df.drop(columns="Footnotes")
ins_contraceptives_df = ins_contraceptives_df.rename(columns={"State Requires Coverage of Prescription Contraception": "Prescription_Contraception", 
                                                             "State Requires Coverage of Over-the-counter Methods": "OTC_Methods",
                                                             "State Requires Coverage of Male Sterilization": "Male_Sterilization",
                                                             "State Requires Coverage of Female Sterilization": "Female_Sterilization",
                                                             "State Prohibits cost sharing": "Cost_Sharing"})

ins_contraceptives_df
```

```python
teen_birth_df = pd.read_csv("Teen_Birth_Rate.csv", 
                           skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72], 
                           index_col='Location')
teen_birth_df = teen_birth_df.drop(columns="Footnotes")
teen_birth_df = teen_birth_df.rename(columns={"Teen Birth Rate per 1,000": "Teen_Births"})
teen_birth_df
```

```python
poverty_df = pd.read_csv("poverty_pop_distribution.csv", 
                           skiprows=[0,1,3,55,56,57,58,59,60,61,62,63,64], 
                           index_col='Location')
poverty_df = poverty_df.drop(columns=["Total", "200-399%", "400%+"])
poverty_df = poverty_df.rename(columns={"Under 100%": "poverty_under_100", "100-199%": "poverty_100_to_199"})
poverty_df
```

```python
add_poverty_df = (poverty_df['poverty_under_100'] + poverty_df['poverty_100_to_199'])
new_poverty_df = pd.DataFrame(data=add_poverty_df, columns=['poverty_under_200'])
new_poverty_df
```

```python
#https://stackoverflow.com/questions/44327999/python-pandas-merge-multiple-dataframes
data_frames = [abortion_status_df, CHC_df, ins_df, maternal_mortality_df, population_df, no_doctor_visit_df, 
              mammogram_df, no_doctor_df, pap_smear_df, ins_contraceptives_df, teen_birth_df, new_poverty_df]
```

```python
from functools import reduce
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Location'],
                                            how='outer'), data_frames)
```

```python
df_merged.index.names = ['State']
```

```python
pd.DataFrame.to_csv(df_merged, 'merged.csv', sep=',', na_rep='.', index=True)
```
